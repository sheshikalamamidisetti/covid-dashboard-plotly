```python
# Install libraries
!pip install plotly pandas --quiet

import pandas as pd
import plotly.express as px

# Load COVID-19 dataset from GitHub
url = "https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv"
df = pd.read_csv(url)

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter countries for example
countries = ['India', 'United States', 'Brazil', 'France', 'Germany']
df = df[df['Country'].isin(countries)]

# 1 Line chart: Daily confirmed cases
fig_line = px.line(df, x='Date', y='Confirmed', color='Country', title='Daily Confirmed COVID-19 Cases')
fig_line.show()

# 2 Bar chart: Total cases by country (latest date)
latest = df[df['Date']==df['Date'].max()]
fig_bar = px.bar(latest, x='Country', y='Confirmed', title='Total Confirmed COVID-19 Cases by Country')
fig_bar.show()

# 3 Pie chart: Total deaths share by country (latest date)
fig_pie = px.pie(latest, names='Country', values='Deaths', title='Total COVID-19 Deaths by Country')
fig_pie.show()
```
